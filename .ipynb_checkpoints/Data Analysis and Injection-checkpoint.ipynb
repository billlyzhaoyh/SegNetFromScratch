```python
def to_byte_tensor(item):
    res = torch.ByteTensor(torch.ByteStorage.from_buffer(item.tobytes()))
    w,h = item.size
    return res.view(h,w,-1).permute(2,0,1)


def to_float_tensor(item): return item.float().div_(255.)
```

```python
#transform into tensors x and y
import torch
x=to_float_tensor(to_byte_tensor(im))
x=x.unsqueeze(dim=0)
y=to_byte_tensor(mk)
```

```python
import torch.nn as nn
import torch.nn.functional as F

class SegNet(nn.Module):
    def __init__(self,data):
        super(SegNet, self).__init__()
        bs,c_in,nh,nw=data.shape
        c_out=64
        kernel_size=7
        c_out_final=13
        conv_pad=int((kernel_size-1)/2)
        self.relu=nn.ReLU()
        #first encoderself.encoder1
        self.encoder1=nn.Conv2d( c_in, c_out, kernel_size,padding=conv_pad)
        self.maxpool1=(nn.MaxPool2d(2, stride=2,return_indices=True))
        #second encoder
        self.encoder2=nn.Conv2d( c_out, c_out, kernel_size,padding=conv_pad)
        self.maxpool2=(nn.MaxPool2d(2, stride=2,return_indices=True))
        #third encoder
        self.encoder3=nn.Conv2d( c_out, c_out, kernel_size,padding=conv_pad)
        self.maxpool3=(nn.MaxPool2d(2, stride=2,return_indices=True))
        #fourth encoder
        self.encoder4=nn.Conv2d( c_out, c_out, kernel_size,padding=conv_pad)
        self.maxpool4=(nn.MaxPool2d(2, stride=2,return_indices=True,ceil_mode=True))
        #first decoder
        self.maxunpool1=(nn.MaxUnpool2d(2, stride=2))
        self.decoder1=nn.Conv2d( c_out, c_out, kernel_size,padding=conv_pad)
        #second decoder
        self.maxunpool2=(nn.MaxUnpool2d(2, stride=2))
        self.decoder2=nn.Conv2d( c_out, c_out, kernel_size,padding=conv_pad)
        #third decoder
        self.maxunpool3=(nn.MaxUnpool2d(2, stride=2))
        self.decoder3=nn.Conv2d( c_out, c_out, kernel_size,padding=conv_pad)
        #fourth decoder
        self.maxunpool4=(nn.MaxUnpool2d(2, stride=2))
        self.decoder4=nn.Conv2d( c_out, c_out_final, kernel_size,padding=conv_pad)
        #final softmax
        self.softmax=torch.nn.Softmax2d()

    def forward(self, x):
        
        x=self.encoder1(x)
        x=self.relu(x)
       
        x,indice_1=self.maxpool1(x)
        
        x=self.encoder2(x)
        x=self.relu(x)
        
        x,indice_2=self.maxpool2(x)
        
        x=self.encoder3(x)
        x=self.relu(x)
        
        x,indice_3=self.maxpool3(x)
       
        x=self.encoder4(x)
        x=self.relu(x)
        
        #store this shape so we can maxunpool to the same shape
        restored_shape=x.shape
    
        x,indice_4=self.maxpool4(x)
        
        
        
        x=self.maxunpool1(x,indice_4,output_size=restored_shape)

        
        x=self.decoder1(x)
        
        x=self.maxunpool2(x,indice_3)
        
        x=self.decoder2(x)
        
        x=self.maxunpool3(x,indice_2)
        
        x=self.decoder3(x)
        
        x=self.maxunpool4(x,indice_1)
        
        x=self.decoder4(x)
        
        
        x=self.softmax(x)
        
        return x
```
